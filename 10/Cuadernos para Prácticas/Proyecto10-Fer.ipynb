```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('diamonds')
```

```python
# Análisis exploratorio
df.head()
```

```python
r1 = sns.relplot(data= df,
            x="price",
            y="carat")
r1.fig.suptitle ("Carat vs Price",va = "center")
r1.set_axis_labels("price [USD]","carat")
```

```python
h1 = sns.displot(data=df,
            x="price",
            kde=True,
            kind="hist")
h1.fig.suptitle("Price histogram")
```

```python
sns.catplot(data = df,
            kind="box",
            x="color",
            y="carat",
           hue="cut")
```

```python
sns.catplot(data = df,
            kind="box",
            x="color",
            y="price",
           hue="cut")
```

```python
# Emplear un jointplot
j1 = sns.jointplot(data=df,
                   x="depth",
                   y="price",
                   kind="kde",
                   fill=True,
                   space=0,
                   color="blue")
```

```python
# Emplear pairplot
p1 = sns.pairplot(data = df,
                 corner = True)
```
