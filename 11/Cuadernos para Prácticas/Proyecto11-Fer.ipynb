```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.model_selection import
```

```python
df = pd.read_csv('../Ventas.csv')
```

```python
df.head(20)
```

```python
df.info()
```

```python
df["Fecha"] = pd.to_datetime(df["Fecha"])
df["Year"] = df["Fecha"].dt.year
df["Month"] = df["Fecha"].dt.month
```

```python
plt.scatter(df["DíaDeLaSemana"],df["Ventas"],color='blue',marker='o')
```

```python
custom_palette = ["#1f77b4",  # Azul
                  "#ff7f0e",  # Naranja
                  "#2ca02c",  # Verde
                  "#d62728",  # Rojo
                  "#9467bd",  # Púrpura
                  "#8c564b",  # Marrón
                  "#e377c2"]  # Rosa
```

```python
sns.relplot(data=df,
            x="Fecha",
            y="Ventas",
            hue="DíaDeLaSemana",
            palette = custom_palette,
            kind="line",
           )
```

```python
# Modelo de regresión lineal
# Aunque este modelo no es el más acertado para predecir las ventas ya que las ventas tienen un comportamiento "sinosoidal"
# Además tienen una tendencia que puede generar una mala estimación de los resultados
fig = plt.figure()
reg = linear_model.LinearRegression()
for idx, iday in enumerate(df["DíaDeLaSemana"].unique()):
    df_tmp = df[df["DíaDeLaSemana"] == iday]
    df_tmp["idx"] = df_tmp.index
    X = df_tmp[["idx"]]
    y =  df_tmp["Ventas"]
    reg.fit(X, y)
    y_predict = reg.predict(X)
    s = reg.score(X,y)
    m = reg.coef_[0]
    b = reg.intercept_

    # Graficos
    plt.plot(df_tmp["Fecha"],y,color=custom_palette[iday-1],label=f"{iday}")
    plt.plot(df_tmp["Fecha"],y_predict,'--', color=custom_palette[iday-1])
    print(f"Dia = {iday} -- score = {s:.4} -- ecc = y={m:.4f}x+{b:.4f}")

plt.legend(title="Dia de la semana",loc="best")
```
