```python
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import LabelEncoder
```

```python
df = pd.DataFrame({
    "Edad":[25, 30, 35, 40, 45],
    "Salario":[3000, 3500, 4000, 4500, 5000]
})
```

```python
f = plt.figure()
p = plt.scatter(df["Edad"],df["Salario"],marker = "o",color="blue")
plt.grid(True)
plt.title("Relaci√≥n entre Edad y Salario")
plt.xlabel("Edad")
plt.ylabel("Salario")
X = df[["Edad"]]
y = df["Salario"]
modelo = linear_model.LinearRegression()
modelo.fit(X,y)
plt.plot(df["Edad"],modelo.predict(X))
```
